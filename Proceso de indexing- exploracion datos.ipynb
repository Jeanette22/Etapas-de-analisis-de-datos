#### Uso de los métodos .loc iloc: también utilizadas para filtrar datos
* recordar: en .LOC colocamos la columna
* recordar: en .ILOC colocamos la posición con numeros de lo que quremos filtrar

```
amazon.loc[:,'dateAdded'] #Cuando usamos el método .LOC se coloca el nombre de la columna, primero las filas
#si quiero todas las filas de tal columna, coloco : 
```

```
amazon.iloc[:,5] #En cambio cuando usamos .ILOC, colocamos la posición de la columa que queremos, y
#espcificamos igual las filas, en este caso solicite lo mismo que lo anterior para corroborar
```

```
amazon.loc [2,'dateAdded'] #en este caso espefico que fila quiero de tal columna con .LOC 
```

```
amazon.iloc [2,5] #tomando el mismo ejemplo de arriba, pero usando .ILOC
```

```
amazon.iloc [0:5, 5] #También con .ILOC  podemos pedir varias filas de una columna especificada 
```

```
amazon.iloc [0:6, 3:6] #en este caso solicite que me muestre las filas que van del 0 al 6, para que me muestre
#hasta la 5, *recordar que el conteo empieza en 0*. Luego coloque que me muestre las columnas que van
#de la 3 a la 6, es decir 3 columnas
```

```
amazon.loc [6,'categories': 'manufacturer'] #mostrame fila especifica de columna mencionada. En este caso usando .LOC!
#puede colocarse el index (revisar ID), en este caso estan representadas con numeros, por eso la aclaracion. Pero
#en otros data frame podemos utilizar algun nombre o dato especifo 
#tambien puede ser dos columnas especficas, se separan con :
```

#### Puede usarse .LOC u .ILOC con condiciones, por ejemplo:

```
reviews_amazon = amazon.loc [amazon["reviews.numHelpful"] < 40]
reviews_amazon
```

```
reviews_amazon.shape #Vemos el total de registros de la condicion mencionada ! 
```

```
#En este caso le agregamos "manufacturer", especificando que solo muestre la condicion especificada 
reviews_amazon = amazon.loc [amazon["reviews.numHelpful"] < 40, ['manufacturer']]
reviews_amazon

```

```
raiting_amazon = amazon.loc [amazon["reviews.rating"] ==3, ['name']]
raiting_amazon
```

```
raiting_amazon.shape #Hay 124 registros con 3 de raiting 
```
